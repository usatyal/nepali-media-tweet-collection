```python
# libraries
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# datasets
annapurnapost = pd.read_csv("annapurnapost.csv", sep=",")
nagariknews = pd.read_csv("nagariknews.csv", sep=",")
republica = pd.read_csv("republica.csv", sep=",")
nayapatrika = pd.read_csv("nayapatrika.csv", sep=",")
setopati = pd.read_csv("setopati.csv", sep=",")
himalkhabar = pd.read_csv("himalkhabar.csv", sep=",")
nepalkhabar = pd.read_csv("nepalkhabar.csv", sep=",")
swasthyakhabar = pd.read_csv("swasthyakhabar.csv", sep=",")
nepallive = pd.read_csv("nepallive.csv", sep=",")
nepallive = pd.read_csv("nepallive.csv", sep=",")
pahilopost = pd.read_csv("pahilopost.csv", sep=",")
kantipur = pd.read_csv("kantipur.csv", sep=",")
thehimalayantimes = pd.read_csv("thehimalayantimes.csv", sep=",")
kathmandupost = pd.read_csv("kathmandupost.csv", sep=",")
```

```python
# combining tweets
words = " ".join(tweet for tweet in thehimalayantimes.Tweet)
# number of words in tweets
print (len(words))
```

```python
# stopword list, will be ignored while building wordcloud
stopwords = set(STOPWORDS)
stopwords.update(["https", "Nepali", "Nepal", "CO", "Horoscope", "Editorial", "country", "people", "THT", "Monday",
"Tuesday", "Saturday", "will", "two", "Friday", "Thursday","Wednesday","Sunday","said","first","one","time","three","today",
                  "opinion","day","year"
                 ])

# wordcloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(words)

# Display the image
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# Save the image in the img folder:
wordcloud.to_file("himalayantimes_wordcloud.png")
```
